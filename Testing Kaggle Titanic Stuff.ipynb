## starting to clean up the data:

```python
# add a new column:
df['Gender'] = 4 # Note - not df.Gender = 4 !!
df
```

## line magic directives:

```python
% lsmagic
```

```python
# define gender column as just M or F
df['Gender'] = df['Sex'].map( lambda x: x[0].upper() )
```

```python
# but actually 0 and 1 is more useful:
df['Gender'] = df['Sex'].map( {'female': 0, 'male': 1} ).astype(int)
```

```python
df['Embarked'].unique()
```

```python
df['Port'] = 4
```

```python
modePort = df['Embarked'].mode()
```

```python
def mapPort(embarked):
    if embarked is None:
      embarked = modePort
    if embarked=='S':
        return 0
    elif embarked=='C':
        return 1
    elif embarked=='Q':
        return 2
    else:
        return modePort
```

```python
df['Port'] = df['Embarked'].map(mapPort)
df
```

```python
# find the median age per gender and class, so that we can use it where the age is unknown.
median_ages = np.zeros((2,3))
median_ages
for i in range(0, 2):
    for j in range(0, 3):
        median_ages[i,j] = df[(df['Gender'] == i) & \
                              (df['Pclass'] == j+1)]['Age'].dropna().median()
median_ages
```

```python
# put this into a new column
df['AgeFilled'] = df['Age']
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFilled']].head(10)
```

```python
for i in range(0, 2):
    for j in range(0, 3):
        df.loc[ (df.Age.isnull()) & (df.Gender == i) & (df.Pclass == j+1),\
                'AgeFilled'] = median_ages[i,j]
df[ df['Age'].isnull() ][['Gender','Pclass','Age','AgeFilled']].head(10)        
```

```python
df['AgeIsNull'] = pd.isnull(df.Age).astype(int)
df.info()
```

```python
# create a new column for family size (number of siblings + number of parents or children on board)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

df['FamilySize'].hist(bins=12,alpha=0.5)
P.show()
```

```python
df.groupby('FamilySize').count().PassengerId
```

```python
# add an age by class column
df['Age*Class'] = df.AgeFilled * df.Pclass
df
```
